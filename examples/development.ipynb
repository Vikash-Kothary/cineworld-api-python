```python
# Imports
import requests
import json
import datetime
import cineworld_api
from pprint import pprint
```

```python
cineworld = cineworld_api.Client()
```

```python
search_term = input('Cinema name:') or 'wembley'
cinema_id = cineworld.search_cinemas_by_name(search_term)['id']
cinema_id
```

```python
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
cinema_dates = cineworld.get_dates(cinema_id)['dates']
for i in range(len(cinema_dates)):
    day = datetime.datetime.strptime(cinema_dates[i], "%Y-%m-%d")
    print('{}: {} {}'.format(i, week[day.weekday()], cinema_dates[i]))
choice = int(input('Enter date:'))
cinema_date = cinema_dates[choice]
#cinema_date
```

```python
# Create schedule
schedule = []
movies = cineworld.get_movies(cinema_id, date=cinema_date)
for movie in movies['films']:
    movie_model = dict()
    movie_model['name'] = movie['name']
    movie_model['running_time'] = movie['length']
    movie_model['id'] = movie['id']
    movie_model['trailer'] = movie['videoLink']
    #movie_model['weight'] = movie['weight']
    schedule.append(movie_model)
    #pprint(movie_model)

for movie in schedule:
    i=0
    for showtime in movies['events']:
        if movie['id'] == showtime['filmId']:
            showtime_model = {
                'show_no': i,
                'datetime': showtime['eventDateTime'],
                'show_type': showtime['attributeIds']
            }
            i+=1
            if showtime['soldOut']:
                showtime_model['soldOut'] = True
            movie.setdefault("showtimes", []).append(showtime_model)
        pprint(showtime)
    
#pprint(schedule)
```

```python
# Remove not interested movies
result = []
for i in range(len(schedule)):
    print('Interested in watching {}? '.format(schedule[i]['name']))
    print(schedule[i]['trailer'])
    interested = input('Y/N:').lower() != 'n'
    if interested:
        schedule[i]['movie_no'] = i
        result.append(schedule[i])
#pprint(result)
```

```python
# Get end times
for movie in result:
    for i in range(len(movie['showtimes'])):
        showtime = movie['showtimes'][i]
        if showtime.get('datetime'):
            starttime = datetime.datetime.strptime(showtime['datetime'], "%Y-%m-%dT%H:%M:%S")
            showtime['startTime'] = starttime.strftime("%H:%M:%S")
            endtime = starttime + datetime.timedelta(minutes=movie['running_time'])
            showtime['timeEnd'] = endtime.strftime("%H:%M:%S")
            showtime.pop('datetime', None)
    #pprint(result[movie])
pprint(result)
```

```python
# Display schedule
def print_schedule(movie, show):
    print('{name} {show_type}: {start} -> {end}'.format(
        name=result[movie]['name'],
        show_type='',
        #show_type=result[movie]['showtimes'][show]['show_type'],
        start=result[movie]['showtimes'][show]['startTime'],
        end=result[movie]['showtimes'][show]['timeEnd']))
```
